```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Generate example data for a 10-class problem
n_samples = 1000
n_classes = 10
np.random.seed(42)  # For reproducibility

# True labels and predicted labels
true_labels = np.random.randint(0, n_classes, size=n_samples)
predicted_labels = np.random.randint(0, n_classes, size=n_samples)

# Generate the confusion matrix
cm = confusion_matrix(true_labels, predicted_labels, labels=np.arange(n_classes))

# Display the confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=np.arange(n_classes))
disp.plot(cmap=plt.cm.Blues, xticks_rotation="vertical")

plt.title("10x10 Confusion Matrix")
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

# Exemplo de labels (substitua com seus dados reais)
true_labels = np.random.randint(0, 10, size=1000)
predicted_labels = np.random.randint(0, 10, size=1000)

# Matriz de confuso
cm = confusion_matrix(true_labels, predicted_labels, labels=np.arange(10))

# Acurcia
accuracy = accuracy_score(true_labels, predicted_labels)

# Sensibilidade (Recall)
recall = recall_score(true_labels, predicted_labels, average="macro")

# Preciso
precision = precision_score(true_labels, predicted_labels, average="macro")

# F-Score
f_score = f1_score(true_labels, predicted_labels, average="macro")

# Especificidade (calculada manualmente para cada classe)
specificity = []
for i in range(len(cm)):
    tn = cm.sum() - (cm[i, :].sum() + cm[:, i].sum() - cm[i, i])
    fp = cm[:, i].sum() - cm[i, i]
    specificity.append(tn / (tn + fp))

specificity = np.mean(specificity)  # Mdia da especificidade para mltiplas classes

# Exibir resultados
print(f"Acurcia: {accuracy:.4f}")
print(f"Sensibilidade (Recall): {recall:.4f}")
print(f"Preciso: {precision:.4f}")
print(f"F-Score: {f_score:.4f}")
print(f"Especificidade (mdia): {specificity:.4f}")
```
