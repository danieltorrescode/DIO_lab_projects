```python
from PIL import Image
import matplotlib.pyplot as plt

image_path = "images/PetImages/Dog/101.jpg"
output_path = "images/grayscale_image.jpg"

# Load and display the image
image = Image.open(image_path)

# Display using Matplotlib
plt.imshow(image)
plt.axis('off')  # Hide axes
plt.show()
```

```python

# Carregar a imagem
image = Image.open(image_path)
# Converter para tons de cinza
grayscale_image = image.convert("L")
# Salvar a imagem em tons de cinza
grayscale_image.save(output_path)
print(f"Imagem convertida para tons de cinza e salva em: {output_path}")
# Load and display the image
image = Image.open(output_path)

# Display using Matplotlib
plt.imshow(image)
plt.axis('off')  # Hide axes
plt.show()
```

```python
output_path = "images/binary_image.jpg"
 # Limite de limiarizao (padro: 128).
threshold=128

# Carregar a imagem
image = Image.open(image_path)
# Converter para tons de cinza se ainda no estiver
grayscale_image = image.convert("L")
# Aplicar binarizao
binary_image = grayscale_image.point(lambda x: 255 if x > threshold else 0, '1')
# Salvar a imagem binarizada
binary_image.save(output_path)
print(f"Imagem binarizada e salva em: {output_path}")
# Load and display the image
image = Image.open(output_path)

# Display using Matplotlib
plt.imshow(image)
plt.axis('off')  # Hide axes
plt.show()
```
